#### Given a square grid of characters in the range ascii[a-z], rearrange elements of each row alphabetically, ascending. Determine if the columns are also in ascending alphabetical order, top to bottom. Return YES if they are or NO if they are not.

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'gridChallenge' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING_ARRAY grid as parameter.
#
def quicksort(lett):
    
    if len(lett) < 2:
        return lett
    from random import randint
    pivot = lett[randint(0,len(lett)-1)]
    low, same, high = [], [], [] 
    
    for j in range(len(lett)):
        if lett[j] < pivot:
            low.append(lett[j])
        elif lett[j] == pivot:
            same.append(lett[j])
        elif lett[j] > pivot:
            high.append(lett[j])
        
    return quicksort(low) + same + quicksort(high)
    
def gridChallenge(grid):
    # Write your code here
    n = len(grid)
    # convert each grid to a list
    new_grid = [list(row) for row in grid]
    for i in range(len(new_grid)):
        new_grid[i] = quicksort(new_grid[i])
    # transpose the new_grid
    transposed = [list(row) for row in zip(*new_grid)]
    
    for row in transposed:
        for i in range(len(row) -1):
            if row[i] > row[i+1]:
                return "NO"
            
    return "YES"
            
        
if __name__ == '__main__':
    # fptr = open(os.environ['OUTPUT_PATH'], 'w')
    # example input
    # 1 -> sample test cases
    # 4 -> number of rows
    # dabc -> row 1
    # lmno -> row 2
    # ghij -> row 3
    # efkp -> row 4
    t = int(input().strip())

    for t_itr in range(t):
        n = int(input().strip())

        grid = []

        for _ in range(n):
            grid_item = input()
            grid.append(grid_item)

        result = gridChallenge(grid)
        print(result)
        # fptr.write(result + '\n')

    # fptr.close()
```

 # Time and Space Complexity

### 1. Quicksort:

**Time Complexity**:

- **Best and Average Case**: $O(n \log n)$, where $n$ is the length of the list.
  
- **Worst Case**: $O(n^2)$. This occurs when the pivot is consistently the smallest or the largest element (though the use of a random pivot reduces the likelihood of hitting the worst-case consistently).

**Space Complexity**: 

- **Average Case**: $O(\log n)$ due to the recursive stack. 
- **Worst Case**: $O(n)$ when the recursion becomes linear.

### 2. gridChallenge function:

a) Converting each string to a list: $O(n \times m)$, where $n$ is the number of strings and $m$ is the length of each string. 

b) Sorting each row using quicksort: $O(n \times m \log m)$ in the average case, and $O(n \times m^2)$ in the worst case.

c) Transposing the grid: $O(n \times m)$.

d) Checking each row in the transposed grid: $O(n \times m)$.

**Total Time Complexity**:

- **Average Case**: $O(n \times m \log m)$.
- **Worst Case**: $O(n \times m^2)$.

**Space Complexity**:

The major space consumers are:

- The space used by the quicksort: $O(m)$ in the worst case.
- The space for `new_grid`: $O(n \times m)$.
- The space for `transposed`: $O(n \times m)$.

Adding up these space complexities, we get:

- **Total Space Complexity**: $O(n \times m) + O(m) = O(n \times m)$.

So, the overall space complexity is $O(n \times m)$.
