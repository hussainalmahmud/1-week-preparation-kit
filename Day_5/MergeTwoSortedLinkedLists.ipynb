```python
#!/bin/python3

import math
import os
import random
import re
import sys

class SinglyLinkedListNode:
    def __init__(self, node_data):
        self.data = node_data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert_node(self, node_data):
        node = SinglyLinkedListNode(node_data)

        if not self.head:
            self.head = node
        else:
            self.tail.next = node


        self.tail = node

def print_singly_linked_list(node, sep, fptr):
    while node:
        fptr.write(str(node.data))

        node = node.next

        if node:
            fptr.write(sep)

# Complete the mergeLists function below.

#
# For your reference:
#
# SinglyLinkedListNode:
#     int data
#     SinglyLinkedListNode next
#
#
def mergeLists(head1, head2):
    dummy = SinglyLinkedListNode(0)
    merged_list = dummy
    curr1 = head1
    curr2 = head2
    
    while curr1 and curr2:
        if curr1.data <= curr2.data: 
            merged_list.next = curr1
            merged_list = merged_list.next
            curr1 = curr1.next
        else:
            merged_list.next = curr2
            merged_list = merged_list.next
            curr2 = curr2.next
    if curr2:
        merged_list.next = curr2        
    else:
        merged_list.next = curr1
        
    return dummy.next
        
if __name__ == '__main__':
    # fptr = open(os.environ['OUTPUT_PATH'], 'w')
    
    # 1 -> no of test cases
    # 4 -> No. of elements in first linked list
    # 1 -> first element in first linked list
    # 4 -> second element in first linked list
    # 5 -> third element in first linked list
    # 8 -> fourth element in first linked list
    # 3 -> No. of elements in second linked list
    # 2 -> first element in second linked list
    # 6 -> second element in second linked list
    # 7 -> third element in second linked list


    tests = int(input())

    for tests_itr in range(tests):
        
        llist1_count = int(input())

        llist1 = SinglyLinkedList()

        for _ in range(llist1_count):
            llist1_item = int(input())
            llist1.insert_node(llist1_item)
            
        llist2_count = int(input())

        llist2 = SinglyLinkedList()

        for _ in range(llist2_count):
            llist2_item = int(input())
            llist2.insert_node(llist2_item)

        llist3 = mergeLists(llist1.head, llist2.head)
        print_singly_linked_list(llist3, ' ', sys.stdout)
        # print_singly_linked_list(llist3, ' ', fptr)
        # fptr.write('\n')

    # fptr.close()
```

 # Time and Space Complexity
 For the `mergeLists` function, let's analyze the time and space complexity:

**Time Complexity:**
1. The main operation in the function is the `while` loop where we compare the nodes of `head1` and `head2` and determine which one should be added to the merged list. In the worst case, this loop runs until one of the linked lists (`head1` or `head2`) is exhausted. So, if `n` is the length of `head1` and `m` is the length of `head2`, the loop runs for `min(n, m)` iterations.
2. After the loop, we append the remaining part of the longer list to the merged list in constant time since we're just updating pointers. This operation takes O(1) time.

Combining these observations, the overall time complexity is O(min(n, m)), but in practical terms and for analysis of algorithms that involve linear iterations through lists or arrays, this is often represented as:
 - #### Time Comlexity `O(n+m)`

**Space Complexity:**
1. We are not using any additional data structures that grow with the size of the input. The `dummy` node and other variables like `merged_list`, `curr1`, and `curr2` take constant space regardless of the input size.
2. The merged list is constructed using the existing nodes from `head1` and `head2`, so we are not allocating new nodes that would contribute to the space complexity.

Given these points, the space complexity is:
 - #### Space complexity `O(1)`
